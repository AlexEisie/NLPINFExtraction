```python
import os
from openai import OpenAI
import pandas as pd
from io import StringIO

os.environ["OPENAI_API_KEY"] = "***"
os.environ["OPENAI_BASE_URL"] = "https://***/v1"
model = 'gpt-4o'

class RESDF():
    def __init__(self,raw_list,batch=5):
        self.df = pd.DataFrame(columns=['Raw Name','Name','Author','Publish'])
        self.batch = batch
        self.names_list = [raw_list[i:i + self.batch] for i in range(0, len(raw_list), self.batch)]
        self.str_list = ['\n'.join(i) for i in self.names_list]

    def nlp_process(self,model,start=0,end=0,save_step=False):
        for i in range(start,len(self.str_list)):
            result = openai_chat(model,self.str_list[i])
            print(result)
            result = pd.read_csv(StringIO('Name,Author,Publish\n'+result))
            result['Raw Name'] = pd.Series(self.names_list[i])
            self.df = pd.concat([self.df,result])
            print(f'Finished {i+1} in {len(self.str_list)}')
            if save_step:
                self.df.to_csv('rear.csv',index=None)
            if end and i >= end:
                print('Reach Limit')
                break

def openai_chat(model,inputs):
    inputs = inputs.replace(",", "，")
    client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"),base_url=os.environ.get("OPENAI_BASE_URL"),)
    completion = client.chat.completions.create(
    model=model,
    messages=[
            {"role": "system", "content": "You are a filename rearrange assistant, I will give you several filename contains 'Series Name', 'Auther', 'Publishing House'(maybe) and others, which are NOT ordered, you need to split out three field I mentioned as csv format and reply csv line directly, no unnecessary contents or column name needed.For Example,'[三坪一魔][鹿島初][Vol.01][青文][电子版]' should be responed with '三坪一魔,鹿島初,青文'."},
            {"role": "user", "content": inputs}
    ])
    return completion.choices[0].message.content.strip('`\n')

def read_raw(filepath):
    with open(filepath,'r') as rf:
        raw_list = rf.read().splitlines()
    return raw_list

if '__main__' == __name__:
    raw_list = read_raw('raw.txt')
    df = RESDF(raw_list,1)
    df.nlp_process(model,0,0,True)
    df.df.to_csv('rear.csv',index=None)
```
